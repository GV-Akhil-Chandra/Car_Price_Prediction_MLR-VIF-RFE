# Car Price Prediction 

### Problem statement
A Chinese automobile company, `Geely Auto`, aspires to enter the US market by setting up their manufacturing unit there and producing cars locally to give competition to their US and European counterparts. 

 

They have contracted an automobile consulting company to understand the factors on which the pricing of cars depends. Specifically, they want to understand the factors affecting car pricing in the American market, as they may differ from the Chinese market. 

The company wants to know the following things:

- Which variables are significant in predicting the price of a car?
- How well do those variables describe the price of a car? 

### Business Goals
We are required to model the price of cars with the available independent variables. The management will use be using this model to understand exactly how the prices vary with the independent variables. Accordingly, they can change the design of the cars, the business strategy, etc., to meet certain price levels. Further, the model will allow the management to understand the pricing dynamics of a new market.

## data Preparation

The variable carbody has five levels. We need to convert these levels into integer. Similarly we need to convert the categorical variables to numeric.

Therefore we will create `dummy variables`.

```python
# dummy variables
significant_cat_col = ['Cars_Category','fueltype','aspiration','carbody','drivewheel','enginetype','cylindernumber']

# makeing dummy variables for the categorical feature
dummies = pd.get_dummies(df[significant_cat_col])
dummies.shape
```

```python
dummies = pd.get_dummies(df[significant_cat_col], drop_first = True)
dummies.shape
```

```python
# Adding the results to original dataframe
df = pd.concat([df, dummies], axis = 1)
```

```python
# Drop the original catagorical variables as the dummies are already created
df.drop(significant_cat_col, axis = 1, inplace = True)
df.shape
```

### Rescaling the Features
In Linear Regression, scaling is a good practice to rescale the variables so that they have a comparable scale. If there are no comparable scales, then some of the coefficients obtained by fitting the regression model might be very large or very small as compared to the other coefficients.<br>
Ways of rescaling:
- Min-Max scaling
- Standardisation (mean-0, sigma-1)

```python
# Using Standardisation Scaling method
scaler = preprocessing.StandardScaler()
```

```python
significant_num_col = ['wheelbase','carlength','carwidth','curbweight','enginesize','boreratio','horsepower','citympg','highwaympg','price']
```

```python
# Apply scaler() to all the columns except the 'dummy' variables
df_train[significant_num_col] = scaler.fit_transform(df_train[significant_num_col])
```

### RFE
Let's use Recursive feature elimination since we have too many independent variables

```python
# Running RFE with an output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)             
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
# Selecting the variables which are in support
col_support = X_train.columns[rfe.support_]
col_support
```

```python
# Creating a new X_train dataframe with RFE selected variables which are in our support
X_train_rfe = X_train[col_support]
```

```python
# Adding a constant variable and Build a first fitted model
# Adding a constant
X_train_rfe_c = sm.add_constant(X_train_rfe)
lm_rfe = sm.OLS(y_train,X_train_rfe_c).fit()

lm_rfe.params
```

```python
#Summary of linear model
print(lm_rfe.summary())
```

### Checking VIF
Variance Inflation Factor in short VIF, gives a basic quantitative idea about how much the feature variables are correlated with each other. It is a very important parameter to test our linear model.

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)

vif
```
